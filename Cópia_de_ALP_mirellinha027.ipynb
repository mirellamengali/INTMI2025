##decimal de PI

```python
from decimal import Decimal, getcontext

def calcular_pi(n):
    # Set a practical limit to avoid MemoryError
    if n > 100000:  # You can adjust this limit based on your environment's capabilities
        print("Warning: The requested number of decimal places is very large and may cause performance issues or errors.")
        print("Calculating with a reduced number of decimal places (e.g., 100000).")
        n = 100000 # Reduce n to a more manageable value

    # Define a precisão (algumas casas extras para evitar erro)
    getcontext().prec = n + 5

    # Algoritmo de Gauss–Legendre
    a = Decimal(1)
    b = Decimal(1) / Decimal(2).sqrt()
    t = Decimal(1) / Decimal(4)
    p = Decimal(1)

    # The number of iterations determines the precision. More iterations for more precision.
    # The original 20 iterations might not be enough for very high precision 'n'.
    # A more dynamic approach based on 'n' might be needed for optimal calculation,
    # but for a fixed practical limit like 100000, a fixed number of iterations might suffice or need tuning.
    # For simplicity with the added limit, we keep a fixed number of iterations.
    for _ in range(20): # This number of iterations might need to be adjusted for higher precision within the limit
        an = (a + b) / 2
        b = (a * b).sqrt()
        t -= p * (a - an) ** 2
        a = an
        p *= 2

    pi = (a + b) ** 2 / (4 * t)
    getcontext().prec = n
    return +pi  # aplica arredondamento

# Entrada do usuário
n_input = int(input("Quantas casas decimais de π você quer calcular? "))
pi = calcular_pi(n_input)
print(f"\nπ com {min(n_input, 100000)} casas decimais:\n{pi}") # Print the actual calculated precision
```

##Crie um algoritimo que receba 2 números e uma escolha que mostre o sequinte menu:

selecione o sinal:

1- Adição

2- Subtração

3- Multiplicação

4- Divisão


caso não escolha um número válido apareça na tela: Opção invalida!

se escolher 1, some dois números e mostre o resultado

se escolher 2, subtraia os dois números e mostre o resultado

e assim com as outras opções

```python
# Entrada dos números
num1 = float(input("Digite o primeiro número:"))
num2 = float(input("Digite o segundo número:"))

# Menu de opções
print("Selecione o sinal:")
print("1 - Adição")
print("2 - Subtração")
print("3 - Multiplicação")
print("4 - Divisão")

# Escolha do usuário
opcao = int(input("Digite a opção desejada (1 a 4): "))

# Estrutura
if opcao == 1:
    resultado = num1 + num2
    print(f"Resultado da adição: {resultado}")
elif opcao == 2:
    resultado = num1 - num2
    print(f"Resultado da subtração: {resultado}")
elif opcao == 3:
    resultado = num1 * num2
    print(f"Resultado da multiplicação: {resultado}")
elif opcao == 4:
    # Evita divisão por zero
    if num2 != 0:
        resultado = num1 / num2
        print(f"Resultado da divisão: {resultado}")
    else:
        print(f"Erro! Não é possível dividir por zero.")
else:
    print(f"Opção inválida!")
```

Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5). Crie um algoritmo em Python que realize esse cálculo corretamente.

```python
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_hora = float(input("Digite o valor por hora: "))

if horas_trabalhadas <= 40:
    salario = horas_trabalhadas * valor_hora
else:
    horas_normais = 40
    horas_extras = horas_trabalhadas - horas_normais
    salario = (horas_normais * valor_hora) + (horas_extras * valor_hora * 1.5)

print(f"O salário total é: R$ {salario:.2f}")
```
