Crie um algaritimo que leia a sua idade e mostre as frases na tela:
Até 12 anos - Voce éuma criança!
13 - 18 - Voce é adolescente!
16 - 60 - Voce é adulto!
Acima de 60 - Voce é idoso!

```python
# Leitura da idade
idade = int(input("Digite sua idade: "))

# Verificação da faixa etária
if idade <= 12:
    print("Você é uma criança!")
elif idade >= 13 and idade <= 18:
    print("Você é adolescente!")
elif idade >= 19 and idade <= 60:
    print("Você é adulto!")
else:
    print("Você é idoso!")
```

Crie um algoritmo que receba 2 números e uma escolha que mostre o seguinte menu:

Selecione o sinal:

1 - Adição

2 - Subtração

3 - Multiplicação

4 - divisão

caso não escolha um número válido apareça na tela: Opção inválida !

se escolher 1 some os 2 números e mostre o resultado.

se escolher 2 subtraia os 2 números e mostre o resultado.

e assim com as outras opções.

```python
# Recebendo os dois números
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

# Mostrando o menu de opções
print("\nSelecione o sinal:")
print("1 - Adição")
print("2 - Subtração")
print("3 - Multiplicação")
print("4 - Divisão")

# Recebendo a escolha do usuário
opcao = input("Digite o número da operação desejada: ")

# Estrutura condicional para escolher a operação
if opcao == "1":
    resultado = num1 + num2
    print(f"Resultado da adição: {resultado}")
elif opcao == "2":
    resultado = num1 - num2
    print(f"Resultado da subtração: {resultado}")
elif opcao == "3":
    resultado = num1 * num2
    print(f"Resultado da multiplicação: {resultado}")
elif opcao == "4":
    # Verifica se o segundo número é zero antes de dividir
    if num2 != 0:
        resultado = num1 / num2
        print(f"Resultado da divisão: {resultado}")
    else:
        print("Erro: Divisão por zero não é permitida!")
else:
    print("Opção inválida!")
```
